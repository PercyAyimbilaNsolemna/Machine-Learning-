## **SPLITTING DATA INTO FEATURES AND TARGET VALUES**

This section splits the training data into training features and target values that is X and y

```python
#Features and converts it to a numpy array
X = data_new.drop(['Churn'], axis=1).values
print(f'The shape of the training data is {X.shape} \n')
print(type(X))

#Target value and converts it to a numpy array
y = data_new['Churn'].values
print(f'The shape of the target variable is {y.shape}')
```

## **SPLITTING THE TRAINING DATA**

This section splits the **data_new into training, cross-validation and testing** data using **train_test_split from scikit learn model selection**

```python
#Splits the data into training, cross-validation and testing
# Get 60% of the dataset as the training set. Put the remaining 40% in temporary variables: x_ and y_.
X_train, X_, y_train, y_ = train_test_split(X, y, train_size=0.6, random_state=42, stratify=y)
# Split the 40% subset above into two: one half for cross validation and the other for the test set
X_test, X_cv, y_test, y_cv = train_test_split(X_, y_, test_size=0.5, random_state=42, stratify=y_)
#Deletes the temporary variables
del X_, y_

print(f'The shape of the training set (input) is {X_train.shape}')
print(f'The shape of the training set (target) is {y_train.shape}\n')
print(f'The shape of the cross validation set (input) is {X_cv.shape}')
print(f'The shape of the cross validation set (target) is {y_cv.shape}\n')
print(f'The shape of the test set (input) is {X_test.shape}')
print(f'The shape of the test set (target) is {y_test.shape}')
```

## **FEATURE SCALING**

This section focuses on feature scaling which enables the model to **converge faster**.

Feature scaling ensures that the features have a **variance and standard deviation of zero**.

**Feature scaling = (feature - mean) / standard_deviation**

```python
#Instantiate the Standard Scaler from scikitlearn
standardScaler = StandardScaler()

#Scales the training set
X_train_scaled = standardScaler.fit_transform(X_train)

#Outputs the calculated mean of the training data
print(f'The calculated mean of the training data for each feature is {standardScaler.mean_}\n')

#Outputs the min and max values in the unscaled X_train data
print(f'The maximum value in the X_train unscaled is {X_train.max()}')
print(f'The minimum value in the X_train unscaled is {X_train.min()}\n')

#Outputs the min and max of the values in the scaled X_train data
print(f'The maximum value in the scaled X_train is {X_train_scaled.max()}')
print(f'The minimum value in the scaled X_train is {X_train_scaled.min()}')
```

```python
#Scales the X_cv using the mean and stadard deviation used to scale the training data
X_cv_scaled = standardScaler.fit_transform(X_cv)

#Scales the X_test using the mean and standard deviation used to scale the training data
X_test_scaled = standardScaler.fit_transform(X_test)
```

## **DEVELOPING THE MODEL**

This section deals with implementing the **Logistic Regression** using libraries from **scikit learn**.

**Different architectures of KNN** will be used to **train the model**. The **models performance** will be evaluated using the **cross validation data set** and finally the architecture with the **best performance** will be used to **preddict the test data set**.

Refer to the other notebook for the same model built using **Neural Network**

```python
#A dictionary to storer the train accuracies
train_accuracies = {}

#A dictionary to store the cross-validation accuracies
cross_validation_accuracies = {}

#Maximum number of iteration for each solver to coonverge
number_of_iterations = np.arange(40, 260, 10)
print(number_of_iterations)
```

```python
#Loops through the number of maximum iterations
for iteration in number_of_iterations:

  #Instantiate the Logistic Regression class and set the random state to 16 so the results can be replicated
  logisticRegression = LogisticRegression(random_state=16, max_iter=iteration)

  #Fit the logistic regression model on the training data
  logisticRegression.fit(X_train_scaled, y_train)

  #Computes accuracy for the train set and adds it to the train accuracies dictionary
  train_accuracies[iteration] = logisticRegression.score(X_train_scaled, y_train)

  #Computes accuracy for the cross-validation set and adds it to the cross validation accuracies dictionary
  cross_validation_accuracies[iteration] = logisticRegression.score(X_cv_scaled, y_cv)

#Outputs all the number of maximum ietrations, train_accuracy scores and the cross validation accuracy scores
print(f'Iterations:                  {number_of_iterations}')
print(f'Train Accuracy:              {train_accuracies}')
print(f'Cross Validation Accuracies: {cross_validation_accuracies}')
```

## **EXPLANATION OF RESULTS**

The results after the accessing the accuracy score for both the train and cross validation after 40 iterations become constant.

This clearly depicts that when **Gradient Descent reaches the global minimum, further iterations do not cause any change in the Mean Squared Error**.

**In order to save resources the model with 50 iterations will be selected as best model in this implementation.**



```python
#Instantiate the Logistic Regression
model = LogisticRegression(random_state=16, max_iter=50)

#Uses the model to fit the scaled training data
model.fit(X_train_scaled, y_train)

#Uses the model to predict the test set
prediction = model.predict(X_test_scaled)

#Outputs the first five Target values and the predicted values
print(f'Target values:   {y_test[:30]}')
print(f'Predited values: {prediction[:30]}')

#Outputs the model's accuracy score on the predicted
accuracy_score = model.score(X_test_scaled, y_test)
accuracy_score = accuracy_score * 100
print(f'\nThe model has an accuracy score of {accuracy_score:.0f}%' )
```
